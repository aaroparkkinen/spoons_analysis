## Results & Findings: Ad Campaign Performance Evaluation

### Description of the campaign

This assessment investigates the impact on performance to a single iOS app in Brazil through an Ad campaign, which ran from 7th of October, 2014 to 31st of December, 2014. This initiative used one Ad creative, and it ran in Facebook app with a daily budget cap of $90.

#### Daily campaign key performance statistics
| Metric | Pre-Campaign | Campaign Value | Impact |
| :--- | :---: | :---: | ---: |
| Daily Downloads | 13 | 2,268 | 174x |
| Daily Organic downloads | 13 | 855 | 65x |
| Daily Net Profit | $1.08 | $111.38 | 103x |

#### Total value created
| Metric | Pre-Campaign | Campaign Value |
| :--- | :---: | :---: |
| Total Net Profit | $72.34 | $9,578.79 |
| Total Cost | $0.00 | $7,563.00 |
| ROAS | N/A | 2.27x |
| ROI | N/A | 126.7% |

These tables clearly demonstrate the success of the campaign in driving visibility, massively increasing downloads and organic user base. It didn't only buy downloads but also achieved sustained growth in the Brazil App Store rankings. By achieving a 126.7% Return on Investment and a 2.27 Return on Ad Spend, the initiative transitioned the app from its baseline organic state to a significant revenue-generating app.

This assessment evaluates the campaign from various perspectives to identify concrete improvements moving forward.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('cleaned_campaign_data.csv')
df['Date'] = pd.to_datetime(df['Date'])
campaign_df = df[df['Date'] >= pd.to_datetime('2014-10-07')].copy()
```

```python
# Plot Showing Daily Cost and Net Profit
plt.figure(figsize = (14, 7))

plt.plot(df['Date'], df['Ad spend'], label = 'Daily Cost')
plt.plot(df['Date'], df['Daily Net Profit'], label = 'Daily Net Profit')

# formatting
plt.axhline(y = 0, color = 'black')
plt.title('Daily Cost vs Net Profit', fontsize = 16)
plt.xlabel('Date')
plt.ylabel('Amount (USD)')
plt.legend()
plt.show()
```

From the Net Profit & Loss plot, we can see the impact of the campaign compared to its cost. The app is profitable throughout the campaign based on pure dollar amounts, however, we can clearly see a strong decline towards break-even levels indicating that the ad campaign might be losing its effectiveness, which we want to look into.

```python
# Chart showing the Average Profit Per Download (APPD)
plt.figure(figsize = (14, 7))

plt.bar(campaign_df['Date'], campaign_df['APPD (on paid downloads)'], 
        color = ['green' if x >= 0 else 'red' for x in campaign_df['APPD (on paid downloads)']])

# formatting
plt.axhline(y = 0, color = 'black')
plt.title('Daily Average Profit Per Download', fontsize = 14)
plt.ylabel('Profit (USD)')
plt.xlabel('Date')
plt.show()
```

**Average profit per download** is a strong metric to assess the performance of the ad campaign, since it specifically focuses on the customers that installed the app through the ad. On days, where APPD goes negative (red), the initiative is losing money. Towards the end of the campaign, this is clearly the case, implying that either the same ad creative has already been seen my majority of users and isn't effective anymore, and that ad visibility during this period is too expensive for the visibility that it is getting. The decreasing effectiveness of the ad was confirmed through analysing click-through rate, which measures the percentage of people clicking the ad after seeing it. Throughout the campaign, the rate was 3.4% on average, but in the last two weeks where it is showing a clear loss, click-through rate fell to 2.7% showing a decreasing interaction by the users for the ad.

**CPM (Cost per 1,000 impressions)** and **CPI (Cost per install)** are crucial performance metrics for the Ad campaign. They explain how much each person seeing the add costs and how much each ad-created install costs. These metrics are especially important since they take the app itself out of the assessment and focuses on the success of the ad itself. Optimally the cost per install should be as low as possible. From the plots below, we can see that while the cost per 1,000 impressions change was only slightly increasing, the cost per install was clearly strongly upward sloping during the campaign, which means that for the same cost, the ad reached almost the same viewership, however, due to the decreased click-through rate, the ad's cost per installing customer became too expensive.

```python
# Plot CPI and CPM trends
sns.set_theme(style = "whitegrid")
fig, axes = plt.subplots(1, 2, figsize = (14, 6))

sns.lineplot(data = campaign_df, x = 'Date', y = 'CPI', ax = axes[0]).set_title('Cost Per Install')
sns.lineplot(data = campaign_df, x = 'Date', y = 'CPM', ax = axes[1]).set_title('Cost Per 1,000 impressions')

# formatting
fig.autofmt_xdate()
plt.tight_layout()
plt.show()
```

### Next steps - How should the app proceed with their marketing?

If the app was to continue their marketing, clear changes have to be made. The ad is losing its effectiveness, the cost for new installs is getting too expensive and the campain itself is losing money for the app. Looking at the APPD and its 7 day moving average, protocols to manage the campaign budget could be adjusted.

During days with clearly positive ad profitability per download, the capped marketing budget is actually limiting the app's success and should be increased. To find the optimal level, futher testing is required but an increase to $180 (2x current budget) is suggested initially on such days with positive APPD level.

To limit the downside, on days with negative profit per download, the ad campaign is losing money and ideally advertisement should be limited to minimal levels. This is difficult in practice but multiple signals assessing current profitability and predicting daily levels could be implemented such that on 'bad', non-profitable advertisement days, budget stays limited, and initial suggested level is set around $20 before further testing.

```python
# Plotting download trends
plt.figure(figsize = (14, 6))

plt.plot(campaign_df['Date'], campaign_df['Paid downloads'], label = 'Paid Downloads', color = 'blue')
plt.plot(campaign_df['Date'], campaign_df['Organic downloads'], label = 'Organic Downloads', color = 'green')

# formatting
plt.title('Paid vs Organic Downloads', fontsize = 14)
plt.ylabel('Daily Downloads')
plt.legend()
plt.grid()
plt.show()
```

#### Possible investments to regain momentum or shutdown the program?

After optimising the advertisement budget on each day for maximal profit, new investments towards the app and a new ad creative should be considered. The decreasing click-through rate indicates a reduction of interest towards the app through. I suggest that since the current ad creative is almos 90 days old, concentrating on a new ad creative, and an effective one through a simple A/B test. The test should focus on whether click-through rate might have declined due to old visuals or rather old messaging.

Before suggesting a potential shutdown, the natural spillover has to be considered. The ad as previously mentioned, doesn't only buy new downloads, but also increases the app's organic user base. The campaign achieved a significant spillover: for every 100 paid downloads, the app gained an additional 60 organic users. This explains why the app is profitable throughout the campaign even when the APPD, the ad's direct profitability metric is negative showing a loss. From the above plot, we can clearly see that the organic downloads were still quite consistent towards the end of the period. Hence, even though most of the metrics are indicating potential shutdown or major changes, there is still an importance on continuous advertisement to maintain these levels.

For the potential shutdown levels, the 7-day moving average line is a suggestion indicating that when it goes negative for seven consecutive days, the ad campaign in its current form is paused or budget majorly limited to the suggested $20 cap. The 7-day moving average describes the trend of the ad and its profitability ignoring singular 'bad' days, giving a more stable perspective on the ad campaign. It is normal for an ad's performance to fluctuate, therefore, singular 'bad' days should be ignored and only perhaps optimised through automated budget adjustments. The ad creative has clearly been exhausted as seen from the negative trend in the plot below, and it shows a sustained loss period. This rule is a suggested protocol on which the campaign in its current form should be shut down to prevent further losses. The shutdown could be permanent or momentary and is always scenario dependent, requiring further assessment. Because the current moving average for APPD has been negative for more than seven consecutive days, I suggest that if the campaign is to be continued, the budget should be immediately limited to a $20 cap to preserve the app's organic momentum while the suggested improvements are put into effect.


```python
# Calculate 7-day moving average trend
campaign_df['7_day_avg'] = campaign_df['APPD (on paid downloads)'].rolling(7).mean()

# Plot APPD and its moving average
sns.set_theme(style = "whitegrid")
plt.figure(figsize = (12, 6))

sns.lineplot(data = campaign_df, x = 'Date', y = 'APPD (on paid downloads)', color = 'gray', alpha = 0.7, label = 'Daily APPD')
sns.lineplot(data = campaign_df, x = 'Date', y = '7_day_avg', color = 'blue', linewidth = 2, label = '7-Day Trend')

# formatting
plt.axhline(y = 0, color = 'black')
plt.title('Ad Profitability Per Download Trend')
plt.show()
```
